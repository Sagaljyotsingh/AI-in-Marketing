```python
import pandas as pd
```

```python
from datetime import datetime, timedelta
```

```python
df = pd.dtFrame(columns=['Date', 'Price', 'Likes', 'Dislikes', 'Followers'])
```

```python
# Define the start and end dates
start_date = datetime(2018, 1, 1)
end_date = datetime(2023, 12, 31)
```

```python
# Define the step (e.g., one day)
step = timedelta(days=1)
```

```python
i = start_date
```

Web Scraping :)

```python
while i <= end_date:
  url = f'https://admn5015-340805.uc.r.appspot.com/{i.date()}.html'
  respose = requests.get(url, timeout=15)
  html = respose.text
  soup = BeautifulSoup(html, "html.parser")
  element = soup.find("td", {"id": "price"})
  if element is not None and element.text is not None:
    price = element.text
  else:
    price = ""
  element = soup.find("td", {"id": "likes"})
  if element is not None and element.text is not None:
    likes = element.text
  else:
    likes = ""
  element = soup.find("td", {"id": "dislikes"})
  if element is not None and element.text is not None:
    dislikes = element.text
  else:
    dislikes = ""
  element = soup.find("td", {"id": "followers"})
  if element is not None and element.text is not None:
    followers = element.text
  else:
    followers = ""
  new_row = {'Date': f'{i.date()}', 'Price': price, 'Likes': likes, 'Dislikes': dislikes, 'Followers': followers}
  print(new_row)
  df = df.append(new_row, ignore_index=True)
  i += step
```

```python
# Create traning and testing dtsets

x = dt[['day','month','year','Likes', 'Dislikes', 'Followers']]
y = dt['Price']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
# Identify the best model

regressors = {
    "XGBRegressor": XGBRegressor(),
    "RandomForestRegressor": RandomForestRegressor(),
    "DecisionTreeRegressor": DecisionTreeRegressor(),
    "GaussianProcessRegressor": GaussianProcessRegressor(),
    "SVR": SVR(),
    "NuSVR": NuSVR(),
    "Ridge":Ridge(),
    "Lars": Lars(),
    "HuberRegressor": HuberRegressor(),
    "PassiveAggressiveRegressor": PassiveAggressiveRegressor(),
    "ARDRegression": ARDRegression(),
    "BayesianRidge": BayesianRidge(),
    "ElasticNet": ElasticNet(),
    "OrthogonalMatchingPursuit": OrthogonalMatchingPursuit(),
}
```

```python
df_models = pd.DataFrame(columns=['model', 'run_time', 'rmse'])

for key in regressors:

    print('*',key)

    start_time = time.time()

    regressor = regressors[key]
    model = regressor.fit(x_train, y_train)
    y_pred = model.predict(x_test)

    scores = cross_val_score(model,
                             x_train,
                             y_train,
                             scoring="neg_mean_squared_error",
                             cv=10)

    row = {'model': key,
           'run_time': format(round((time.time() - start_time)/60,2)),
           'rmse': round(np.sqrt(mean_squared_error(y_test, y_pred)))
    }

    df_models = pd.concat([df_models, pd.DataFrame.from_records([row])])
```

```python
# Order model results

df_models.head(20).sort_values(by='rmse', ascending=True)
```

The best regressor is Random Forest Regressor, as it has the lowest RMSE value of 88.
